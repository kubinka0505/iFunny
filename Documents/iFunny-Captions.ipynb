```
#@title # <font color=red>⬅</font> **3.** Fill the form <font color="orange">**AND THEN**</font> press ▶ to process and display `Image`
#@markdown ---

#@markdown # `Text`
Content = "kubinka0505" #@param {type:"string"}
Additional_Wrap = -8 #@param {type:"integer"}
Kerning = -1 #@param {type:"integer"}

#@markdown #`Media`
#@markdown ## `Image`
URL = "https://kubinka0505.github.io/assets/media/img/png/Banners/Main.png" #@param {type:"string"}
Scale_Back = False #@param {type:"boolean"}
Watermark = False #@param {type:"boolean"}
#@markdown ## `Video`
#@markdown ### `Duration`
Start = 0 #@param {type:"number"}
End = 0 #@param {type:"number"}
#@markdown ### `Audio`
Direct_URL = "https://kubinka0505.github.io/assets/media/sound/mp3/Piano/1.mp3" #@param {type:"string"}
Bitrate = 192 #@param {type:"slider", min:8, max:320, step:8}
Channels = "2 (Stereo)" #@param ["1 (Mono)", "2 (Stereo)", "3 (Left → Mono)", "4 (Right → Mono)"] {type:"string"}
Channels = Channels[0]
#@markdown # `Settings`
#@markdown ## `Font`
Type = "2 (Futura Condensed Extra Bold)" #@param ["1 (Roboto Condensed Bold)", "2 (Futura Condensed Extra Bold)"] {type:"string"}
Type = Type.split(" ")[0]
#@markdown
Delay = 5 #@param {type:"number"}
Dark_Mode = False #@param {type:"boolean"}
Logs = "0 (Default)" #@param ["-1 (Silent)", "0 (Default)", "1 (Verbose)"]
Logs = Logs.split(" ")[0]

#@markdown ## `Caption_Design`
#@markdown ### `Percentage_Elements_Size`
Font = 10 #@param {type:"number"}
Emojis = 67 #@param {type:"number"}
Leading = 112 #@param {type:"number"}
Caption_Field_Height = 150 #@param {type:"number"}
Design = ",".join([str(Value) for Value in (Font, Emojis, Leading, Caption_Field_Height)])

#@markdown ### `Colors`
Text = "#000" #@param {type:"string"}
Caption_Field = "#FFF" #@param {type:"string"}
Colors = ",".join([str(Value) for Value in (Text, Caption_Field)])

#@markdown ---
#@markdown ## ℹ️ **Displaying image can take a moment!**
Show_Help = False #@param {type:"boolean"}
Preview_Video_Volume = 50 #@param {type:"slider", min:0, max:100, step:1}
Preview_Video_Volume = str(Preview_Video_Volume / 100)

#---#

from IPython.display import *
from base64 import b64encode as b64e
from google.colab import files, output
chdir("/content/iFunny-Captions")

#---#

environ["TEXT"] = Content
environ["WRAP"] = str(Additional_Wrap)
environ["KERNING"] = str(Kerning)
environ["IMAGE"] = URL
environ["START"] = str(Start)
environ["END"] = str(End)
environ["SCALE_BACK"] = "-sb" if Scale_Back else ""
environ["WATERMARK"] = "-wm" if Watermark else ""
environ["AUDIO"] = Direct_URL
environ["BITRATE"] = str(Bitrate)
environ["CHANNELS"] = Channels
environ["FONT"] = Type
environ["DELAY"] = str(Delay)
environ["DESIGN"] = Design
environ["COLORS"] = Colors
environ["DARK_MODE"] = "-dm" if Dark_Mode else ""
environ["VERBOSE"] = "-v " + Logs
environ["HELP"] = "-h" if Show_Help else ""
environ["OUTPUT"] = "Images"

!python3 __init__.pyw \
	-t "$TEXT" \
	-w $WRAP \
	-k $KERNING \
	-i "$IMAGE" \
	-ls $START \
	-le $END \
	-a "$AUDIO" \
	-ab $BITRATE \
	-ac $CHANNELS \
	$SCALE_BACK \
	$WATERMARK \
	-f $FONT \
	-d $DELAY \
	-pes $DESIGN \
	-c $COLORS \
	$DARK_MODE \
	$VERBOSE \
	$HELP \
	-o $OUTPUT \

#---#

chdir("/content/iFunny-Captions/" + environ["OUTPUT"])
Latest_File = path.abspath(max(listdir(), key = path.getctime))

#---#

def Display(File_Path: path, width: int) -> Image:
	"""Displays file using iPython's `display` function."""
	Format = File_Path.split(sep)[-1].split(".")[-1]
	Video = Format not in ("png", "gif")
	#---#
	with open(File_Path, "rb") as File:
		return display(HTML('<{0} src="data:{1}/{2};base64,{3}" width={4}>'.format(
			'video controls onloadstart="this.volume({0})"'.format(
				Preview_Video_Volume
			) if Video else "img",
			"video" if Video else "image",
			Format,
		b64e(File.read()).decode("ascii"),
		width
			)
		)
	)

#---#

if not Show_Help:
	display(HTML("""
		<hr><br>
		<code>Displaying latest generated file</code>
		<br>
		"""
		)
	)

	print()
	Display(Latest_File, width = 338)
	print()
	files.download(Latest_File)
```
