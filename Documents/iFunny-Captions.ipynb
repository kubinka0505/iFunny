```
#@title # <font color=red>⬅</font> **1.** Install
#@markdown Packages:
#@markdown - ## <a href="https://github.com/ffmpeg/ffmpeg" target=_blank>`FFmpeg`</a> >= 4.2.0
#@markdown - ## <a href="https://github.com/kohler/gifsicle" target=_blank>`Gifsicle`</a>
#@markdown - ## <a href="https://github.com/kornelski/pngquant" target=_blank>`PNGQuant`</a>
#@markdown ---

#---#

from os import *
del open

Commands = [
	"add-apt-repository -y ppa:jonathonf/ffmpeg-4",
	"apt-get update",
	"apt-get install ffmpeg gifsicle pngquant"
]

for Command in Commands:
	environ["COMMAND"] = Command
	!$COMMAND
```

```
#@title # <font color=red>⬅</font> **3.** Fill the form <font color="orange">**AND THEN**</font> press ▶ to process and display `Image`
#@markdown ---

#@markdown ### `Text`Content = "text def 123 :snake:" #@param {type:"string"}
Content = "kubinka0505" #@param {type:"string"}
Additional_Wrap = 3 #@param {type:"integer"}
Kerning = -1 #@param {type:"integer"}

#@markdown ### `Image`
Image_URL = "https://avatars.githubusercontent.com/u/30776937" #@param {type:"string"}
Scale_Back = False #@param {type:"boolean"}

#@markdown ### `Font`
Type = 2 #@param ["1", "2"] {type:"raw"}

#@markdown ## `Settings`
Delay = 5 #@param {type:"number"}
Packages_Logs = False #@param {type:"boolean"}

#@markdown ## `Percentage_Elements_Size`
Font = 10 #@param {type:"number"}
Emojis = 67 #@param {type:"number"}
Leading = 112 #@param {type:"number"}
Caption_Field_Height = 150 #@param {type:"number"}
Emoji_Height_Spacing = 25 #@param {type:"number"}
Design = ",".join([str(x) for x in (Font, Emojis, Leading, Caption_Field_Height, Emoji_Height_Spacing)])

#@markdown ### `Dark Mode`
After_Hour = "Disabled" #@param ["Disabled", "00:00", "01:00", "02:00", "03:00", "04:00", "05:00", "06:00", "07:00", "08:00", "09:00", "10:00", "11:00", "12:00", "13:00", "14:00", "15:00", "16:00", "17:00", "18:00", "19:00", "20:00", "21:00", "22:00", "23:00"]

#@markdown ---
#@markdown ## ℹ️ **Displaying image can take a moment!**
Show_Help = False #@param {type:"boolean"}

#---#

from IPython.display import *
from base64 import b64encode as b64e
from google.colab import files, output
chdir("/content/iFunny-Captions")

#---#

Dark_Mode_Enabled = 0
Dark_Mode_Hour = 0
if not After_Hour == "Disabled":
	Dark_Mode_Enabled = 1
	Dark_Mode_Hour = str(int(After_Hour.split(":")[0]))

#---#

environ["TEXT"] = Content
environ["WRAP"] = str(Additional_Wrap)
environ["KERNING"] = str(Kerning)
environ["IMAGE"] = Image_URL
environ["SCALE_BACK"] = "-sb" if Scale_Back else ""
environ["FONT"] = str(Type)
environ["DELAY"] = str(Delay)
environ["DESIGN"] = str(Design)
environ["DARK_MODE"] = ("-dm " + Dark_Mode_Hour) if Dark_Mode_Enabled else ""
environ["VERBOSE"] = "-v" if Packages_Logs else ""
environ["HELP"] = "-h" if Show_Help else ""

!python3 __init__.pyw \
	-t "$TEXT" \
	-w $WRAP \
	-k $KERNING \
	-i "$IMAGE" \
	$SCALE_BACK \
	-f $FONT \
	-d $DELAY \
	-pes $DESIGN \
	$DARK_MODE \
	$VERBOSE \
	$HELP \
	-nop

chdir("/content/iFunny-Captions/Images")
Latest_File = path.abspath(max(listdir(), key = path.getctime))

#---#

def Display(File_Path: path, width: int) -> Image:
	"""Displays image in using iPython's `display` function."""
	with open(File_Path, "rb") as File:
		return display(HTML('<img src="data:image/{0};base64,{1}" width={2}>'.format(
			"gif" if File_Path.split(sep)[-1].split(".")[-1] == "gif" else "png", 
			b64e(File.read()).decode("ascii"), width)))

def _Download(): return files.download(Latest_File)

#---#

if not Show_Help:
	display(HTML("""
		<style>
		#img {
			width: 1em;
			margin-right: .5em
		}

		#dl {
			display: inline-flex;
			align-items: center;
			justify-content: center;
			margin: 0 auto;
			width: 6.75em;
			height: 2em;
			color: #FFCC00;
			background: black;
			cursor: pointer;
			font-weight: bold;
			border-radius: .25em;
			outline: 2px inset #FFCC00
		}
		</style>
		<hr><br>
		<div>
			<code>Displaying latest Image File</code>&#x3000;
			<span id="dl">
				<img id="img" src="https://raw.githubusercontent.com/kubinka0505/iFunny-Captions/master/Documents/Pictures/Main/iFunny.svg">
				Download
			</span>
		</div>
		<script>
		document.querySelector("#dl").onclick = () => {
			google.colab.kernel.invokeFunction("dl", [], {})
		}
		</script>
		<br>
		"""
		)
	)

	print()
	Display(Latest_File, width = 338)
	print() 

	output.register_callback("dl", _Download)
```
